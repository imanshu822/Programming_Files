```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y1 = [50,40,70,80,20]
y2 = [80,20,20,50,60]
y3 = [70,20,60,40,60]
y4 = [80,20,20,50,60]

print("UID : 21BCS5996\nANSHU")
plt.plot(x, y1,'r', label='AUDI', linewidth=2)
plt.plot(x, y2,'c', label='BMW', linewidth=3)
plt.plot(x, y3,'k', label='JAGUAR', linewidth=1)
plt.plot(x, y4,'y', label='LAMBORGHINI', linewidth=2)

plt.title('Car details')
plt.ylabel('Distance in kms')
plt.xlabel('Days')

plt.legend()

plt.show()
```
